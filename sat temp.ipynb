```python
t0 = 350.0

c1 = 73.649	
c2 = -7258.2
c3 = -7.3037
c4 = 4.17E-06
c5 = 2

tolerance = 1e-08

# ---------- NEWTON RAPHSON -------------------------
def func(t,p):
    ans = p - np.exp(c1 + c2/t + c3*np.log(t) + c4*np.power(t,c5))
    return ans

def dev_func(t):
    dev_ans = (-np.exp(c1 + c2/t + c3*np.log(t) + c4*np.power(t,c5))) * (-c2/(t*t) + c3/t + c4*c5*t)
    return dev_ans
# ----------------------------------------------------
```
