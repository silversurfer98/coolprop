```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
```

```python
x = tf.Variable(3.0)

with tf.GradientTape() as tape:
    y = x**2 + 10

tape.gradient(y,x)
```

```python
x = [5,6,7,8,9,10,11,12,13]
m = tf.Variable(1.0)
b = tf.Variable(0.0)
#x = np.array(x, dtype='float32')

with tf.GradientTape() as tape:
    y = m*x + b
    l = tf.reduce_mean(y**2)

[m,b] = tape.gradient(l,[m,b])

y = m*x + b
plt.plot(x,y)
```
